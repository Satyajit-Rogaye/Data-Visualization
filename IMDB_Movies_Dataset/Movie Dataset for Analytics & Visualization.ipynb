# Movie Dataset for Analytics & Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('movies_dataset.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

# Univarite

# Bar & Count plot

```python
genre=data['Genre'].value_counts().index
genre_value=data['Genre'].value_counts().values
```

```python
plt.bar(genre,genre_value,color='red',label='count of genre')
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Counts')
plt.legend()
plt.title("Counts of Genre")
plt.grid()
plt.show()
```

```python
sns.countplot(x='Genre',data=data,order=data['Genre'].value_counts().index,color='orange')
plt.xticks(rotation=90)
plt.show()
```

# Pie Chart

```python
country=data['Country'].value_counts()[:4]
```

```python
plt.figure(figsize=(5,10))
plt.pie(country,labels=country.index,autopct="%1.1f%%",explode=(0.1,0,0,0),startangle=90)
plt.title("Top 4 country where movies are highly released")
plt.show()
```

# Histogram

```python
plt.hist(data['ReleaseYear'],edgecolor='black',label="Year VS Counts")
plt.xlabel("Year")
plt.ylabel("Movies Counts")
plt.title("Total movies released in years")
plt.legend()
plt.grid()
plt.show()
```

# Bar Chart

```python
Top_5_director=data['Director'].value_counts().head().index
Top_5_director_values=data['Director'].value_counts().head().values
Top_5_leadac=data['LeadActor'].value_counts().head().index
Top_5_leadac_values=data['LeadActor'].value_counts().head().values
```

```python
Top_5_director,Top_5_director_values,Top_5_leadac,Top_5_leadac_values
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
# Director graph
axs[0].bar(Top_5_director, Top_5_director_values, color='green', label='Director released movie')
axs[0].tick_params(axis='x', rotation=90)
axs[0].set_xlabel("Director")
axs[0].set_ylabel("Total movies")
axs[0].grid()
axs[0].legend()
axs[0].set_title("Total movies released by each director")

# Lead Actor graph
axs[1].bar(Top_5_leadac, Top_5_leadac_values, color='blue', label='Lead Actor in Movies')
axs[1].tick_params(axis='x', rotation=90)
axs[1].set_xlabel("Lead Actor")
axs[1].set_ylabel("Total movies")
axs[1].grid()
axs[1].legend()
axs[1].set_title("Total Lead Actor in movies")

plt.tight_layout()
plt.show()
```

```python
yearly_sales=data.groupby('ReleaseYear')['US_BoxOfficeUSD'].mean().reset_index()
yearly_sales_global=data.groupby('ReleaseYear')['Global_BoxOfficeUSD'].mean().reset_index()
fig,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].plot(yearly_sales['ReleaseYear'],yearly_sales['US_BoxOfficeUSD'],color='orange',marker='.',label='US Box Offic USD')
ax[0].set_xlabel("Release Year")
ax[0].set_ylabel("US Box Office USD")
ax[0].grid()
ax[0].legend()
ax[0].set_title("US box office sales by year")
ax[1].plot(yearly_sales_global['ReleaseYear'],yearly_sales_global['Global_BoxOfficeUSD'],color='green',marker='.',label='Global Box Office USD')
ax[1].set_xlabel('Release Year')
ax[1].set_ylabel('Global Office USD')
ax[1].grid()
ax[1].legend()
ax[1].set_title('Global office sales by year')
plt.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].scatter(yearly_sales['ReleaseYear'],yearly_sales['US_BoxOfficeUSD'],color='orange',label='US Box Office USD')
ax[0].set_xlabel("Release Year")
ax[0].set_ylabel("US Box Office USD")
ax[0].grid()
ax[0].legend()
ax[0].set_title("US box office sales by year")
ax[1].scatter(yearly_sales_global['ReleaseYear'],yearly_sales_global['Global_BoxOfficeUSD'],color='green',label='Global Box Office USD')
ax[1].set_xlabel('Release Year')
ax[1].set_ylabel('Global Office USD')
ax[1].grid()
ax[1].legend()
ax[1].set_title('Global office sales by year')
plt.tight_layout()
plt.show()
```

```python
top_5_Dire=data['Director'].value_counts().index[:5]
top_3_Genre=data['Genre'].value_counts().index[:3]
top_5_data=data.loc[(data['Director'].isin(top_5_Dire))&(data['Genre'].isin(top_3_Genre))]
plt.figure(figsize=(10,7))
sns.countplot(x='Director',data=top_5_data,hue='Genre')
plt.title("Director's Heat Their Movie in particular Genre")
plt.grid()
plt.show()
```

```python
stack_data=pd.crosstab(index=top_5_data['Director'],columns=top_5_data['Genre'])
stack_data
```

```python
stack_data.plot(kind='bar',stacked=True)
plt.title("Counts of genre as per Director")
plt.ylabel("Counts")
plt.grid()
plt.show()
```

```python
genre_avg_rating=data.groupby('Genre')['IMDbRating'].mean().sort_values(ascending=False)
genre_avg_rating
```

```python
plt.figure(figsize=(8,5))
genre_avg_rating.plot(kind='bar',color='skyblue',label='IMDbRating')
plt.xlabel("Genre")
plt.ylabel("IMDbRating")
plt.grid()
plt.legend()
plt.title("IMDbRating for Genre")
plt.show()
```

```python
country_sales=data.groupby('Country')['Global_BoxOfficeUSD'].mean().sort_values(ascending=False)
US_sales=data.groupby('Country')['US_BoxOfficeUSD'].mean().sort_values(ascending=False)
country_sales
```

```python
plt.figure(figsize=(8,5))
country_sales.plot(kind='bar',color='Yellow',label='Global Sales')
plt.xlabel("Country")
plt.ylabel("Global Sales")
plt.grid()
plt.legend()
plt.title("Country's Global Sales")
plt.show()
US_sales.plot(kind='bar',color='red',label='US Sales')
plt.xlabel("Country")
plt.ylabel("US Sales")
plt.grid()
plt.legend()
plt.title("Country's US Sales")
plt.show()
```

```python
top_10_director=data['Director'].value_counts().index[:10]
top_3_con=data['Country'].value_counts().index[:3]
new_con_dir_data=data.loc[(data['Director'].isin(top_10_director))(data['Country'].isin(top_3_con))]
new_con_dir_data
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Director',data=new_con_dir_data,hue='Country')
plt.xticks(rotation=90)
plt.title("Director's Movies Heat In Top 3 Country")
plt.grid()
plt.show()
```

```python
new_stack_data=pd.crosstab(index=new_con_dir_data['Director'],columns=new_con_dir_data['Country'])
new_stack_data.plot(kind='bar',stacked=True)
plt.grid()
plt.ylabel("Counts")
plt.title("Director's Movies Heat In Top 3 Country")
plt.show()
```
